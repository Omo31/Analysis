This code counts how many times each tag was used.

```python
mostused = dict() # mostviewed = {} will also give the same result

for tags in question["Tags"]:
    for tag in tags:
        if tag in mostused:
            mostused[tag] += 1
        else:
            mostused[tag] = 1
mostused       
```

```python
mostuse = {} 

for tags in allquestion["Tags"]:
    for tag in tags:
        if tag in mostuse:
            mostuse[tag] += 1
        else:
            mostuse[tag] = 1
mostuse        
```

```python
tagquest = pd.DataFrame.from_dict(mostused, orient="index")
tagquest.rename(columns={0: "count"}, inplace=True)
tagquest.head(10)
```

This code shows how many times a tag was viewed

```python
tagview = {}

for index, row in question.iterrows():
    for tag in row['Tags']:
        if tag in tagview:
            tagview[tag] += row['ViewCount']
        else:
            tagview[tag] = row['ViewCount']
tagview
```

```python
tagview = pd.DataFrame.from_dict(tagview, orient="index")
tagview.rename(columns={0: "mostviewed"}, inplace=True)
tagview
```

```python
tagview.sort_values(ascending=True, by="mostviewed").tail(20).plot(kind="barh", figsize=(8, 8))
```

# Combining the two dataframe to further explore the project

```python
inused = pd.merge(tagquest, tagview, how="left", left_index=True, right_index=True)
inused
```

```python
inviewed = pd.merge(tagview, tagquest, how="right", left_index=True, right_index=True)
inviewed
```

```python
alltag = list(tagquest.index)
```

```python
associations = pd.DataFrame(index=alltag, columns=alltag)
associations.iloc[0:4,0:4]
```

```python
associations.fillna(0, inplace=True)

for tags in question["Tags"]:
    associations.loc[tags] += 1
    
```

```python
relations_most_used = associations.loc[inused.index, inused.index]

def style_cells(x):
    helper_df = pd.DataFrame('', index=x.index, columns=x.columns)
    helper_df.loc["time-series", "r"] = "background-color: yellow"
    helper_df.loc["r", "time-series"] = "background-color: yellow"
    for k in range(helper_df.shape[0]):
        helper_df.iloc[k,k] = "color: blue"
    
    return helper_df

relations_most_used.style.apply(style_cells, axis=None)
```
