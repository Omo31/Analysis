# Analysing the Datasets

```python
matches['winner'].value_counts().plot(kind='bar')
```

```python
deliveries['batting_team'].value_counts().plot(kind='bar')
```

```python
deliveries['bowling_team'].value_counts().plot(kind='bar')
```

```python
confactors = matches.pivot_table(values=['win_by_runs', 'win_by_wickets'], index='winner', aggfunc='sum', margins=True)
confactors = confactors.drop('All')
confactors
```

```python
confactors.plot(kind='bar', title='Factors Contributing to Win/Loss', legend=True, figsize=(12,8))
```

# Conclusion

From the analysis above, 

The most successful teams of the matches are Mumbai Indians, Chennai Super Kings, Royal Challengers Bangalore and Kolkata Knight Riders.	

The most successful players are CH Gayle, AB de Villiers, RG Sharma, DA Warner, MS Dhoni

The factor that contributed to their winning is that they won mostly by runs.


From the result of the analysis i suggest that companies should endorse the following teams Mumbai Indians, Chennai Super Kings, Royal Challengers Bangalore and Kolkata Knight Riders and CH Gayle, AB de Villiers, RG Sharma, DA Warner, MS Dhoni for its products.
