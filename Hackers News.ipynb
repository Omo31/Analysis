# Hacker News.

Hacker News is a site started by the startup incubator Y Combinator, where user-submitted stories (known as "posts") are voted and commented upon, similar to reddit. Hacker News is extremely popular in technology and startup circles, and posts that make it to the top of Hacker News' listings can get hundreds of thousands of visitors as a result.

```python
hn = pd.read_csv("hacker_news.csv")
pd.options.display.max_columns = 150 # to avoid truncated output 
hn.head(5)
```

The code above displays the first five rows of hackers news

```python
import re
```

```python
askpost = hn[hn["title"].str.contains("^Ask Hn", flags=re.I)]
askpost.reset_index(inplace=True)
askpost
```

Ask post shows rows that starts with ask post

```python
hn.shape
```

```python
showpost = hn[hn["title"].str.contains("^Show Hn", flags=re.I)]
showpost.reset_index(inplace=True)
showpost
```

Show post shows rows that starts with show post

```python
other = hn.drop(hn[hn["title"].str.contains("^Ask Hn", flags=re.I)].index)
other.reset_index(inplace=True)
other
```

```python
otherpost = other.drop(other[other["title"].str.contains("^Show Hn", flags=re.I)].index)
otherpost.reset_index(inplace=True)
otherpost
```

Other post shows rows that did not starts with ask post and show post

```python
askpost["num_comments"].dtype
```

```python
ask_tot_comments = askpost["num_comments"].sum()
print(ask_tot_comments)
ask_avg_comments = askpost["num_comments"].mean()
print(ask_avg_comments)
```

Ask post received a total comments of 24483 and the average comment for ask post is 14.038

```python
show_tot_comments = showpost["num_comments"].sum()
print(show_tot_comments)
show_avg_comments = showpost["num_comments"].mean()
print(show_avg_comments)
```

Show post received a total comments of 11988 and the average comment for show post is 10.31

```python
import datetime as dt
```

```python
cols = ["created_at", "num_comments"]
ask_result = askpost[cols]
ask_result
```

```python
ask_result["created_at"] = pd.to_datetime(ask_result["created_at"]).dt.strftime('%H')
ask_result["created_at"]
```

```python
ask_result
```

```python
hour = dict(ask_result.pivot_table(index="created_at", values="num_comments", aggfunc=np.sum))
hour
```

This output shows the total number of comments ask post received per hour

```python
avg_hour = dict(ask_result.pivot_table(index="created_at", values="num_comments", aggfunc=np.mean))
avg_hour
```

This output shows the average number of comments ask post received per hour

# CONCLUSION

From this analysis we observed that averagely, ask post received more comments[14.038] than show post[10.31].
Based on this analysis, we conclude that to maximize the amount of comments a post receives, we'd recommend the post be categorized as ask post and created between 15:00 and 16:00, seeing that ask post created at ths time received the highest average comments.
